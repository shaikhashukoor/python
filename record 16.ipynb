```python
class bank:
    def __init__(self,name,ac,typeac,bal,):
        self.name=name
        self.ac=ac
        self.typeac=typeac
        self.bal=bal
        
    def deposit(self,amount):
        self.bal+=amount
        print("deposited amount",amount)
        print("current balance",self.bal)
    
    def withdrew(self):
        self.am=int(input("withdrew amount"))
        print("current balance",self.bal)
        if self.am > self.bal:
            print("you dont have sufficient account balance")
            print("current balance",self.bal)
        else:
            print(self.am,"is withdrewned")
            self.bal-=self.am
            print("current balance",self.bal)
    def accountinfo(self):
        print("account holder name: ",self.name)
        print("account number: ",self.ac)
        print("account type: ",self.typeac)
        print("account balance: ",self.bal)

name=input("enter the name")
ac=int(input("account number"))
typeac=input("type of account")
bal=int(input("balance in account"))
holder=bank(name,ac,typeac,bal)

while(1):
    ch=int(input("1. deposit \n 2. withdrew \n 3. account info \n 4. exit \n CHOOSE YOUR OPTION:"))
    if ch==1:
        damount=int(input("deposit amount:"))
        holder.deposit(damount)
    elif ch==2:
        holder.withdrew()
    elif ch==3:
        holder.accountinfo()
    else:
        print("exiting")
        break



        
        
            
```
