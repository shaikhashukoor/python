```python
#to copy odd lines of one file to other
filea=input("enter the source file:")
fileb=input("enter the destination file:")
f1=open(filea,"r")
f2=open(fileb,"w")
i=0
for line in f1.readlines():
    i=i+1
    if i%2==0:
        continue
    f2.write(line)
f1.close()
f2.close()
print("odd lines copied succesfully from", filea,"to",fileb)
```

```python
#to read each row from a given csv file and print a list of strings
import csv
name=input("enter csv file name:")
f1=open(name,"r")
reader=csv.reader(f1)
for row in reader:
    print(row)
f1.close()
```

```python
#to read speceific column from a given csv file and print the content of the columnc
import csv
fname=input("enter csv file name:")
with open(fname,"r")as f1:
    reader=csv.DictReader(f1)
    print("\n DEPARTMENT DETAILS:")
    print("------------------------")
    for row in reader:
        print("DeptId:",row['DeptId'],"\t DeptName:",row['DeptName']) 
```

```python
#to wite a python dict to csv file. after read the csv file abd display 
import csv
data = [ {"Name": "Anu", "Age": 22, "Marks": 85},
         {"Name": "Arun", "Age": 21, "Marks": 78},
         {"Name": "nefla", "Age": 22, "Marks": 89}]
with open("studentsdict.csv", "w", newline='') as file:
    fieldnames=["Name", "Age", "Marks"]
    writer = csv.DictWriter(file,fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
print("dictnory data written succesfully")
with open("studentsdict.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
